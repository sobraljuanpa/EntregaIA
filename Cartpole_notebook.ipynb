# Simulación aleatoria con n muestreos

```python
def simulate_random_agent(number_of_samples):
    rewards = [0] * number_of_samples #inicializo lista para almacenar recompensas
    for index in range(number_of_samples):
        obs = env.reset()
        done = False
        while not done:
            policy = randint(0,1)
            obs, reward, done, info = env.step(policy)
            rewards[index] += reward
            if done:
                env.reset()
    env.close()
    average = np.average(rewards)

    return average, rewards

print("Simulacion aleatoria 10 muestreos")
start = time.time()
print("Valor esperado :")
print(simulate_random_agent(10))
end = time.time()
print("------- Timetaken(seconds) ----------------")
print(end-start)

print("Simulacion aleatoria 100 muestreos")
start = time.time()
print("Valor esperado :")
print(simulate_random_agent(100))
end = time.time()
print("------- Timetaken(seconds) ----------------")
print(end-start)

print("Simulacion aleatoria 1000 muestreos")
start = time.time()
print("Valor esperado :")
print(simulate_random_agent(1000))
end = time.time()
print("------- Timetaken(seconds) ----------------")
print(end-start)

print("Simulacion aleatoria 10000 muestreos")
start = time.time()
print("Valor esperado :")
print(simulate_random_agent(10000))
end = time.time()
print("------- Timetaken(seconds) ----------------")
print(end-start)

```

# Comenzando a entrenar nuestro agente

```python
q_values = np.zeros((96, 2))
#inicializo apoyandome en la funcion de numpy, estan los 96 estados posibles y las 2 acciones posibles para cada estado
def get_next_action(location, epsilon):
    if np.random.random() < epsilon:
        return np.argmax(q_values[location])
    else:
        return randint(0,1)

def get_location_index(location):
    if location < 0:
        return np.floor((location * 10) - 1).astype(int)
    else:
        return np.floor((location*10) + 48).astype(int)

#esta funcion no va a servir para nada hasta estar entrenada
#def get_best_action(location):

epsilon = 0.9 #90% de las veces tomo la mejor accion y no una aleatoria
discount_factor = 0.9 #no pienso demasiado en recompensas futuras
learning_rate = 0.9 #esta anda a explicarla

#vamo a hace 10 episodios por ahora
for episode in range(10):
    obs = env.reset() # obs es array forma [location, speed, ang(radianes), speed angular]
    done = False
    while not done:
        # elijo accion a tomar
        location = get_location_index(obs[0])
        print(location)
        action = get_next_action(location, epsilon)
        
        # realizo accion, calculo dif temporal
        obs, reward, done, info = env.step(action)
        old_q_value = q_values[location,action]
        temporal_difference = reward + (discount_factor * np.max(q_values[location])) - old_q_value

        # actualizo q value
        new_q_value = old_q_value + (learning_rate * temporal_difference)
        q_values[location, action] = new_q_value

print("Entrenamiento completado")
print(q_values)




```
