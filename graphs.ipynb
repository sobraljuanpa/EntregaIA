```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

import gym
import time
from random import randint
import math
```

```python
def graph_distribution(min_val, max_val):
    NUMBER_OF_BINS = 10
    position_bins = np.linspace(min_val, max_val, NUMBER_OF_BINS)
    obtained_values = np.zeros(NUMBER_OF_BINS + 1, dtype=int)

    env = gym.make('CartPole-v1')
    for episode in range(50000):
        obs = env.reset()
        done = False
        while not done:
            policy = randint(0,1)
            obs, reward, done, info = env.step(policy)
            obtained_bin = np.digitize(obs[0], position_bins)
            obtained_values[obtained_bin] += 1
            if done:
                env.reset()
    env.close()
    #print(obtained_values)
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    bins = np.arange(NUMBER_OF_BINS + 1)
    ax.bar(bins, obtained_values)
    plt.show()

graph_distribution(-4.8, 4.8)
graph_distribution(-2.4, 2.4)
graph_distribution(-0.25, 0.25)
graph_distribution(-0.2, 0.2)
```
